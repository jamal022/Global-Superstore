```python
# here as we can see in column [Order Date , Ship Date ] in Object data type which can be in datetime datatype for better proccessing
# So we are going to create a function to do that task for us
def convert_to_datetime(df,feature):
    df[feature] = pd.to_datetime(df[feature])
```

```python
date_list = ['Order Date','Ship Date']

for feature in date_list:
    convert_to_datetime(df,feature)
```

```python
df.dtypes
```

```python
# After convertig into Datetime objects, convert into a Series objects
Order_date = pd.Series(df['Order Date'], name='Order_Date')
Ship_date = pd.Series(df['Ship Date'], name='Ship_Date')
```

```python
df.columns.values
```

```python
df.head()
```

```python
# Now Let's Drop unwanted columns 
df.drop(columns=['Row ID', 'Order Date', 'Ship Date', 'Customer ID', 'Postal Code'], inplace=True)
df.head(n=2)
```

```python
# Concat new columns to the original data
global_data = pd.concat([df, Order_date, Ship_date], axis=1)
global_data.head(n=2)
```

```python
# Reindex the data to be in the original order
global_data = global_data.reindex(columns=['Order ID', 'Order_Date', 'Ship_Date', 'Ship Mode','Customer Name', 'Segment', 'Postal Code', 'City','State', 'Country', 'Region', 'Market', 'Product ID', 'Category',
       'Sub-Category', 'Product Name', 'Sales', 'Quantity', 'Discount', 'Profit', 'Shipping Cost', 'Order Priority'])
```
